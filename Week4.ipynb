```python
for item in Highlights:
    Name_work = item.find('span', class_='slide-caption').text
    description = item.find('span', class_='slide-description').text
    link = 'https://www.nga.gov/features/black-artists.html#art' + item.find('a')['href']
    print(Name_work, description, link)
```

```python
Artists_Works = []
About = []
Pictures= []

for item in Highlights:
    Name_work = item.find('span', class_='slide-caption').text
    description = item.find('span', class_='slide-description').text
    link = 'https://www.nga.gov/features/black-artists.html#art' + item.find('a')['href']
    
    Artists_Works.append(Name_work)
    About.append(description)
    Pictures.append(link)
```

```python
import pandas as pd
import csv
```

```python
Artists_Works
```

```python
df = pd.DataFrame(
    {'Artist_Works': Artists_Works,
     'About': About,
     'Pictures': Pictures,
    })
```
