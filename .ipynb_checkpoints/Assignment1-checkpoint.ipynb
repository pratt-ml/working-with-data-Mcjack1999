```python
#removing () in strings to avoid errors 

grammar = ['ArtistBio', 'Nationality', 'BeginDate', 'EndDate', 'Gender']

pattern = r'\(|\)'

df_artworks[grammar] = df_artworks[grammar].replace(pattern, '', regex=True)
```

During cleaning I decided to uniform the date values by changing any with multiple years, ie "1997-78", to the first date presented. This will help clean the column for easier analyzes later. While doing so I noticed that some rows had multiple observations saved to it. I proceed to explode this date into their own rows and delete ConstiuteID duplicates, this will help with dataset merging later. 

Based on the low percentage of NA values, I have decided to keep NA values and focus on cleaning the data set for datatypes

I have decided to change date forms with multiple dates in the (i.e. 1976-77) to present the number that comes first in the range to avoid too many NaT outputs. If I didn't do this the NaT outputs for date range exceeded 30%. I will also be keeping the 0s in the end date column to represent that the artist is still alive by replacing it will placeholder '9999-12-31'.

My last step is to remove all syntax errors and double check datatypes and null percentage.

```python
df_artworks['EndDate'] = df_artworks['EndDate'].replace('0', '9999-12-31')
```

```python
import re
import pandas as pd

def extract_first_year(date_string):
    if isinstance(date_string, str):
        # Check for date range like '1976-77'
        range_match = re.match(r'(\d{4})-\d{2}', date_string)
        if range_match:
            return range_match.group(1)  # Return the first year as string
        
     
        year_match = re.match(r'(\d{4})', date_string)
        if year_match:
            return year_match.group(1)  # Return the year if it's a normal year format
    return date_string  

df_artworks['Date'] = df_artworks['Date'].apply(extract_first_year)
```

```python
datefixer = ['Date', 'DateAcquired', 'BeginDate', 'EndDate']

df_artworks[datefixer] = df_artworks[datefixer].apply(pd.to_datetime,
                                              format='%Y',
                                              errors='coerce')
```
