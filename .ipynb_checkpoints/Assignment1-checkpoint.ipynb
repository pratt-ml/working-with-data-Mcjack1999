Exploratory Analysis and Cleaning of the Artwork DataFrame

During the cleaning process, I standardized the date values by selecting the first year from any entries with date ranges (e.g., “1977-78”). This helped clean the date column, making it easier to analyze later. While working through the data, I noticed that some rows contained multiple observations. My initial plan was to explode these into individual rows and remove duplicate ConstituentIDs to facilitate dataset merging. However, due to memory limitations, I had to set this plan aside for now.

As an alternative, I chose to remove all duplicate rows. Initially, I retained rows with NA values due to their low percentage and focused on cleaning datatypes. However, for consistency, I ultimately decided to drop all rows containing NA values.

To handle date ranges (e.g., “1976-77”), I opted to keep only the first year in the range to reduce the number of NaT (Not a Time) outputs. Without this approach, NaT values would have exceeded 30%. I also replaced 0s in the ‘EndDate’ column with ‘2025-12-31’ as a placeholder to indicate that the artist is still alive.

I initially planned to convert all date columns to datetime format. However, I encountered issues with Pandas reverting previously cleaned strings. To resolve this, I filtered for four-digit years, appended ‘-01-01’ to those entries, and replaced 0s in the ‘EndDate’ column with ‘2025’.

For the ‘BeginDate’ column, I treated 0s as missing data (NaN), as they represent an unknown birth year, which I consider equivalent to “unknown.”

Cleaning the artists’ dataset was more straightforward, though I did find inconsistencies in how gender was represented. I standardized the gender values for uniformity.

```python
#I realized that the ‘ConstituentID’ column must contain only a single unique value per row since it serves as a key linking each artist to their respective artworks. 
#I treated it as a primary key (like in databases), ensuring that each artwork is accurately attributed to one artist based on this unique identifier.

df_artworks.loc[:, 'ConstituentID'] = df_artworks['ConstituentID'].str.strip()

# Filter rows where 'ConstituentID' does not contain commas
df_artworks = df_artworks[~df_artworks['ConstituentID'].str.contains(',', na=False)]

# Remove duplicate 'ConstituentID' rows
df_artworks = df_artworks.drop_duplicates(subset=['ConstituentID'], keep=False)
```

```python
#This is where I cleaned the date column to only show full years 
import re

def clean_date_column(date_string):
    if isinstance(date_string, str):
        # Search for a 4-digit year in the string (e.g., from 'Before 1951', 'c. 1995')
        year_match = re.search(r'(\d{4})', date_string)
        if year_match:
            return year_match.group(1)  # Return only the year as a string
    
    return date_string  # Return the original value if no year is found

df_artworks['Date'] = df_artworks['Date'].apply(clean_date_column)
```

```python
def filter_four_digit_year(date_string):
    # Check if it's a string and contains exactly 4 digits
    if isinstance(date_string, str) and re.fullmatch(r'\d{4}', date_string):
        return date_string
    else:
        return np.nan  # Replace anything that is not a 4-digit year with NaN

# Apply the function to the 'Date' column
df_artworks['Date'] = df_artworks['Date'].apply(filter_four_digit_year)
```

```python
def extract_year_as_datetime(date_string):
    if isinstance(date_string, str):
        # Check if the string is a valid 4-digit year
        if re.match(r'^\d{4}$', date_string):
            return pd.to_datetime(date_string, format='%Y')  # Convert to datetime
    return np.nan  # Return NaN for invalid or non-year values

# Apply the cleaning function to 'Date', 'BeginDate', and 'EndDate'
columns_to_convert = ['Date', 'BeginDate', 'EndDate']

for column in columns_to_convert:
    # Keep only valid 4-digit years and convert to datetime
    df_artworks[column] = df_artworks[column].apply(extract_year_as_datetime)
```

```python
# Defined a dictionary to map the current values to standardized values
gender_mapping = {
    'male': 'male',
    'female': 'female',
    'unknown. (non-binary or trans?)': np.nan,  # Replace unknown with NaN
    'gender non-conforming': 'non-binary',
    'non-binary': 'non-binary',
    'transgender woman': 'female (transwoman)',  # Change to 'female (transwoman)'
    'female (transwoman)': 'female (transwoman)'  # Keep 'female (transwoman)' as is
}

# Used replace to clean up the 'Gender' column
df_artists['Gender'] = df_artists['Gender'].replace(gender_mapping)

print(df_artists['Gender'].unique())
```
